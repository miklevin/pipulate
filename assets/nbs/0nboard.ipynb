### The Memory Bridge

In a standard notebook, when you close the tab, the variables disappear. In a web application, state is preserved as a user moves from page to page. 

Pipulate bridges this gap using the `wand`. Let's set a target URL for our first analysis and save it to the local database.

```python
# Step 2: Define the target
TARGET_URL = "https://example.com"

wand.set("onboarding_job", "target_url", TARGET_URL)
print(f"Target saved to database: {TARGET_URL}")
```

**Restart the Kernel**. Press the `Esc` key, then type `0`, `0`. 

Normally, this would wipe the notebook's memory entirely. Run the next cell to see how the `wand` recovers your state. This mechanism is the foundation for turning these explicit, top-to-bottom scripts into robust web applications.

```python
from pipulate import wand

recovered_url = wand.get("onboarding_job", "target_url")
print(f"State recovered. Target is: {recovered_url}")
```

### Giving the Machine Eyes

We have a voice, and we have persistent memory. Now, we need optics.

To prepare the web for an AI, we cannot simply extract text. We must translate the DOM into a format a Large Language Model can comprehend. Because you are running Pipulate locally, we can make this process visible. 

**Run the next cell. Keep your hands off the mouse and watch the machine work.**

```python
# Step 3: Execute the scrape

# Set this to True if you want to force the browser to open and pull fresh data
FORCE_FRESH_SCRAPE = False 

wand.speak(f"Initializing browser automation for {recovered_url}.")

# headless=False means the browser window will manifest on your screen (if not cached).
result = await wand.scrape(
    url=recovered_url, 
    take_screenshot=True, 
    headless=False,
    override_cache=FORCE_FRESH_SCRAPE,
    verbose=True
)

if result.get('success'):
    if result.get('cached'):
        print("⚡ Cache Hit! Using existing artifacts. Change FORCE_FRESH_SCRAPE=True to override.")
    else:
        print("✅ Fresh Scrape Successful.")
else:
    wand.speak("I encountered an error during navigation.")
    print(f"Scrape Failed: {result.get('error')}")
```

### The Side Effects

The browser opened, evaluated the page, and closed. In its wake, it dropped a series of artifacts into a specific folder on your local drive (`browser_cache/example.com/`).
